## Long-form data vs. Wide-from Data

### Long-form data

```python
flights = sns.load_dataset('flights')
```

```python
flights.shape
```

```python
flights.sample(10)
```

```python
sns.relplot(
    data=flights,
    x='year',
    y='passengers',
    hue='month',
    kind='line'
);
```

### Wide-form data

```python
flights_wide = flights.pivot(index='year', columns='month', values='passengers')
```

```python
flights_wide.sample(10)
```

```python
sns.relplot(
    data=flights_wide,
    kind='line'
);
```

### Long-form data vs. Wide-form data

```python
sns.relplot(
    data=flights,
    x='month',
    y='passengers',
    hue='year',
    kind='line'
);
```

```python
sns.relplot(
    data=flights_wide.T,
    kind='line'
);
```
