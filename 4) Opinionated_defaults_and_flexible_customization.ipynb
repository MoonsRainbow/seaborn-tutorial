### relplot() - 자동 생성되는 범례보기

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins
```

```python
# legend='auto'
sns.relplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
);
```

### 플롯 꾸미기

```python
plt.ioff()
```

```python
sns.set_theme(
    style='ticks',
    font_scale=1.25
)
```

```python
grid = sns.relplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
    palette='crest',
    marker='x',
    s=100
)
```

```python
grid.set_axis_labels('Bill Length (mm)', 'Bill Depth (mm)', labelpad=10)
```

```python
grid.legend.set_title('Body Mass (g)')
```

```python
grid.figure.set_size_inches(6.5, 4.5)
```

```python
grid.ax.margins(.15)
```

```python
grid.despine(trim=True)
```

```python
plt.show()
```
