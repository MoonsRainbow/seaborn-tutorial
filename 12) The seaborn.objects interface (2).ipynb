## Transforming data before plotting
: 플로팅 전 데이터 변환

### Statistical transformation
: 통계적 변환

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.sample(10)
```

```python
plot = so.Plot(
    penguins,
    x='species',
    y='body_mass_g'
)
```

```python
plot.add(
    so.Bar(),
    so.Agg()
)
```

```python
plot.add(
    so.Dot(pointsize=10),
    so.Agg()
)
```

```python
plot = so.Plot(
    penguins,
    x='species',
    y='body_mass_g',
    color='sex'
)
```

```python
plot.add(
    so.Dot(pointsize=10),
    so.Agg()
)
```

### Resolving overplotting

```python
plot.add(
    so.Bar(),
    so.Agg()
)
```

```python
plot.add(
    so.Bar(),
    so.Agg(),
    so.Dodge()
)
```

```python
sns.barplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='sex',
    errorbar=None,
    dodge=False
);
```

```python
sns.barplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='sex',
    errorbar=None,
    dodge=True
);
```

```python
plot.add(
    so.Dot(),
    so.Dodge()
)
```

```python
plot.add(
    so.Dot(),
    so.Dodge(),
    so.Jitter(.3)
)
```

### Creating variables through transformation

```python
plot = so.Plot(
    penguins,
    x='species'
)
```

```python
plot.add(
    so.Bar(),
    so.Hist()
)
```

```python
plot = so.Plot(
    penguins,
    x='flipper_length_mm'
)
```

```python
plot.add(
    so.Bars(),
    so.Hist()
)
```

```python
plot = so.Plot(
    penguins,
    x='body_mass_g',
    y='species',
    color='sex'
)
```

```python
plot_var = plot.add(
    so.Range(),
    so.Est(errorbar='sd'),
    so.Dodge()
)
```

```python
plot_var = plot_var.add(
    so.Dot(),
    so.Agg(),
    so.Dodge()
)
```

```python
plot_var
```

### Orienting marks and transforms

```python
plot.add(
    so.Bar(),
    so.Agg(),
    so.Dodge()
)
```

```python
tips = sns.load_dataset('tips')
```
