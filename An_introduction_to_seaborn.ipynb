### 예제 데이터 셋 불러오기

```python
tips = sns.load_dataset('tips')
tips
```

### 시각화

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    hue='smoker',
    style='smoker',
    size='size'
);
```
