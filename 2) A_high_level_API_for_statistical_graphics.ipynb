## dots 데이터 받아오기

```python
dots = sns.load_dataset('dots')
dots
```

## relplot(kind='line') 시각화

```python
sns.relplot(
    data=dots, 
    kind='line',
    x='time', 
    y='firing_rate', 
    col='align',
    hue='choice', 
    style='choice',
    size='coherence', 
    facet_kws=dict(sharex=False),
);
```

# Statistical estimation
: 통계적 추정

## fmri 예제 데이터 받아오기

```python
fmri = sns.load_dataset('fmri')
fmri
```

## relplot() - 통계적 추정, 신뢰 구간, 불확실성 오차 막대 시각화

```python
sns.relplot(
    data=fmri, 
    kind='line',
    x='timepoint', 
    y='signal', 
    col='region',
    hue='event', 
    style='event',
);
```

## lmplot() - 선형 회귀 모델 시각화

```python
sns.lmplot(
    data=tips, 
    x='total_bill', 
    y='tip', 
    col='time', 
    hue='smoker'
);
```

# Distributional representations
: 분포 표현

## displot() - 변수 분포 시각화, 히스토그램과 kde 시각화


```python
sns.displot(
    data=tips,
    x='total_bill',
    col='time',
    kde=True
);
```

```python
sns.displot(
    data=tips,
    x='total_bill',
    col='time'
);
```

## displot() - 경험적 데이터의 누적 분포(ecdf), rug 시각화

```python
sns.displot(
    data=tips,
    kind='ecdf',
    x='total_bill',
    col='time',
    hue='smoker',
    rug=True
);
```

# Plots for categorical data
: 범주형 데이터에 대한 플롯

## catplot() - 군집 분포 시각화

```python
sns.catplot(
    data=tips,
    kind='swarm',
    x='day',
    y='total_bill',
    hue='smoker'
);
```

## catplot() - 바이올린 플롯 시각화

```python
sns.catplot(
    data=tips,
    kind='violin',
    x='day',
    y='total_bill',
    hue='smoker',
    split=True
);
```

## catplot() - 바 플롯 시각화

```python
sns.catplot(
    data=tips,
    kind='bar',
    x='day',
    y='total_bill',
    hue='smoker'
);
```
