## Figure-level vs. axes-level functions

### 모듈 불러오기

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### 백엔드 설정

```python
plt.ion()
```

### displot(kind='hist')
- displot()의 kind 파라미터의 기본값은 'hist'

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    multiple='stack'
);
```

### displot(kind='kde')

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    multiple='stack',
    kind='kde'
);
```

### displot(kind='hist', col='species')

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    col='species'
);
```

### plt.subplots()와 Seaborn의 축(axes) 수준 함수들

```python
# 객체 지향형 시각화 코드 작성을 위한 백엔드 변경
plt.ioff()
```

```python
figure, axes = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(8, 4),
    gridspec_kw=dict(width_ratios=[4, 3])
)
```

```python
sns.scatterplot(
    data=penguins,
    x='flipper_length_mm',
    y='bill_length_mm',
    hue='species',
    ax=axes[0]
)
```

```python
sns.histplot(
    data=penguins,
    x='species',
    hue='species',
    shrink=.8,
    alpha=.8,
    legend=False,
    ax=axes[1]
)
```

```python
figure.tight_layout()
```

```python
plt.show()
```
